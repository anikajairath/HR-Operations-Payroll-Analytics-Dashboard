```python
#Employee count by location
employee['Location'].value_counts().sort_values(ascending=False)
```

```python
#Employee count by Department
employee['Department'].value_counts().sort_values(ascending=False)
```

```python
#Employee Status
employee['Status'].value_counts()
```

```python
#Employees joining by year
employee['Year'] = employee['JoiningDate'].dt.year
employee.groupby('Year')['EmployeeID'].count()
```

```python
#Employees by Department
employee.groupby('Department')['EmployeeID'].count().sort_values(ascending=False)
```

```python
#Creating new column Deduction Percentage
payroll['Deduction Percentage'] = (payroll['Deductions']/payroll['GrossSalary']*100)
payroll['Deduction Percentage'].head()
```

```python
#Creating new column Salary Band
def salary_band(net_pay):
    if net_pay < 60000:
        return "Low"
    elif net_pay <= 100000:
        return "Medium"
    else:
        return "High"

payroll['Salary Band'] = payroll['NetPay'].apply(salary_band)
payroll['Salary Band'].value_counts().sort_values(ascending=False)
```

```python
#Claim by months
expense['Claim Month'] = expense['ClaimDate'].dt.month
expense['Claim Month'].value_counts().sort_values(ascending=False)
```

```python
#Expense by Claim Type
expense.groupby('ClaimType')['Amount'].sum().sort_values(ascending=False)
```

```python
#Expense status overview
expense['Status'].value_counts()
```

```python
#Monthly expense
expense.groupby('Claim Month')['Amount'].sum()
```

```python
#Months with highest expense
expense.groupby('Claim Month')['Amount'].sum().sort_values(ascending=False).head(5)
```

```python
#Travel by mode
travel['Mode'].value_counts()
```

```python
#Travel expense by mode
travel.groupby('Mode')['Cost'].sum()
```

```python
#Top 5 cities most travelled to
travel['ToCity'].value_counts().sort_values(ascending=False).head(5)
```

```python
#Monthly travel trend
travel['TravelMonth'] = travel['TravelDate'].dt.month
travel.groupby('TravelMonth')['Cost'].sum()
```

```python
#Highest payroll depts

merged = pd.merge(employee, payroll, on='EmployeeID')
merged.groupby('Department')['NetPay'].sum().sort_values(ascending=False).head(5)
```

```python
employee.to_excel("employees_cleaned.xlsx", index=False)

payroll.to_excel("payroll_cleaned.xlsx", index=False)

expense.to_excel("expense_cleaned.xlsx", index=False)

travel.to_excel("travel_cleaned.xlsx", index=False)
```
